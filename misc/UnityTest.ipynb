# Drift velocity and its dependence with the electric field

```python
import numpy as np

def vd_e(mu_e,vSat, E):
    
    return (mu_e*E/np.sqrt(1 + (mu_e*E/vSat)**2))

def vd_h(mu_h,vSat, E):
    return (mu_h*E/(1 + (mu_h*E/vSat)))

E = np.arange(0,6e4)
```

## Silicon 

```python
import matplotlib.pyplot as plt
mu_e = 1450 # cm/Vs
mu_h = 450 # cm/Vs
vSat = 0.8e07 #cm/s

vsat_list = [vSat]*len(E)
plt.xlabel("Electric field (V/cm)")
plt.ylabel("Drif velocity (cm/s)")
plt.plot(E,vd_e(mu_e,vSat,E),label="e$^-$")
plt.plot(E,vd_h(mu_h,vSat,E),label="hole")
plt.plot(E,vsat_list)
plt.legend()
```
